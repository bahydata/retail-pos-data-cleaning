## Detected Issues

- Missing values in: contact_no, amt_paid, customer_name, total_amt, unitPrice, receipt_no, sale_date
- Duplicates: there are 9 rows duplicated
- Inconsistent text: Columns show inconsistent formats and casing: mixed date formats in sale_date, inconsistent capitalization in branch names,    and inconsistent casing in pay_status values.

- Data type issues: Columns require type normalization: sale_date → date, and qty, unitPrice, total_amt, amt_paid → numeric.

## Cleaning Decisions

- Drop duplicated rows based on receipt_no,item_desc
- Handle missing values
- Normalize text (strip + lower + title)
- Convert data types

```python
for col in df.select_dtypes(include="object"):
    df[col] = df[col].str.strip().str.lower().str.title()

df.head()
```

```python
df.columns = df.columns.str.strip().str.lower()
df.head()
```

```python
df=df.drop_duplicates(subset=["receipt_no","item_desc"])
df.duplicated().sum()
```

```python
# Convert dates
df["sale_date"] = pd.to_datetime(df["sale_date"],format="mixed",dayfirst=True, errors="coerce")
# Convert numeric
df["qty"]=df["qty"].replace(r"[^\d.]","",regex=True).pipe(pd.to_numeric)
df["unitprice"]=df["unitprice"].replace(r"[^\d.]","",regex=True).pipe(pd.to_numeric)
df["total_amt"]=df["total_amt"].replace(r"[^\d.]","",regex=True).pipe(pd.to_numeric)
df["amt_paid"]=df["amt_paid"].replace(r"[^\d.]","",regex=True).pipe(pd.to_numeric)
```

```python
# Handle with Outliers + Consistency Rules

# recompute derived
df["total_amt"] = df["qty"] * df["unitprice"]

# remove invalid negatives
df = df[(df[['qty','unitprice','total_amt','amt_paid']] >= 0).all(axis=1)]

# enforce logical constraints
df = df[df["total_amt"] >= df["amt_paid"]]

# amount → status
df.loc[df['amt_paid'] == 0, 'pay_status'] = 'Pending'
df.loc[df['amt_paid'] == df['total_amt'], 'pay_status'] = 'Paid'
df.loc[(df['amt_paid'] > 0) & (df['amt_paid'] < df['total_amt']), 'pay_status'] = 'Partial'

# status → amount
df.loc[df["pay_status"].isin(["Paid","Cash"]), "amt_paid"] = df["total_amt"]
df.loc[df["pay_status"] == "Pending", "amt_paid"] = 0

# edge case
df.loc[df["qty"] == 0, "total_amt"] = 0
```

```python
#Handle with NAN values
df[['sale_date','customer_name','contact_no']] = (
    df.groupby('receipt_no')[['sale_date','customer_name','contact_no']]
      .ffill()
      .bfill()
)
df=df.dropna(subset=["receipt_no"])
```

```python
print(df.isna().sum()[lambda x: x!=0].sort_values(ascending=False))
print(df.info())
df.shape
```

```python
# === DATA VALIDATION REPORT ===

# === CONFIG (EDIT PER PROJECT) ===
KEY = ['receipt_no','item_desc']
NON_NEGATIVE_COLS = ['qty','unitprice','total_amt','amt_paid']

DERIVED_RULES = {
    'total_amt': lambda df: df['total_amt'] == (df['qty'] * df['unitprice'])
}

print("=== DATA QUALITY SUMMARY ===")

# Shape
print(f"\nDataset size: {df.shape[0]} rows × {df.shape[1]} columns")

# Missing
missing = df.isna().sum()
missing = missing[missing != 0].sort_values(ascending=False)

print("\nMissing Values:")
if missing.empty:
    print("No missing values")
else:
    print(missing.to_string())

# 🔴 ASSERT — Missing
assert missing.empty, "Missing values still exist"

# KEY checks
print("\nKey Integrity:")

key_nan = df[KEY].isna().sum().sum()
print("Missing in key columns:",
      "No" if key_nan == 0 else f"Yes ({key_nan})")

# 🔴 ASSERT — Key missing
assert key_nan == 0, "Missing values in KEY columns"

key_dups = df.duplicated(subset=KEY).sum()
print("Duplicate records (based on key):",
      "No" if key_dups == 0 else f"Yes ({key_dups})")

# 🔴 ASSERT — Duplicates
assert key_dups == 0, "Duplicate records still exist"

# Value rules
print("\nValue Validation:")

if NON_NEGATIVE_COLS:
    neg_check = (df[NON_NEGATIVE_COLS] >= 0).all().all()
    print("No negative values in critical fields:",
          "Yes" if neg_check else "No")

    # 🔴 ASSERT — Negative values
    assert neg_check, "Negative values detected"

# 🔴 ASSERT — Overpayment
assert (df['amt_paid'] <= df['total_amt']).all(), "Overpayment detected"

# 🔴 ASSERT — Payment rule (المهم)
assert not (
    (df['pay_status'].isin(['Paid','Cash'])) &
    (df['amt_paid'] != df['total_amt'])
).any(), "Payment rule violated"

# Derived checks
print("\nCalculated Fields Consistency:")

for col, rule in DERIVED_RULES.items():
    valid = rule(df).all()
    print(f"{col} is consistent:",
          "Yes" if valid else "No")

    # 🔴 ASSERT — Derived fields
    assert valid, f"{col} calculation is inconsistent"

# Sample
print("\nSample Preview:")
display(df.head(3))
```

```python
# === AFTER CLEANING (Quick Snapshot) ===

print(f"Shape: {df.shape[0]} rows × {df.shape[1]} cols")

print("\nMissing :")
display(df.isna().sum()[lambda x: x!=0].sort_values(ascending=False))

print("\nDuplicates (based on KEY):",
      df.duplicated(subset=KEY).sum())

print("\nSample:")
display(df.head(3))
```
